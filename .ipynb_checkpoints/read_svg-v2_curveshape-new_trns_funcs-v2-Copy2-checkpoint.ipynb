# parse svg to get control points and length for each curve

```python
from svg.path import parse_path
import xml.etree.ElementTree as ET
import numpy as np

ns = "{http://www.w3.org/2000/svg}"
tree = ET.parse("../bear_detached2.svg")
root = tree.getroot()

g = root.findall(ns + "g")[0]
paths = g.findall(ns  + "path")
path_objs = []

for p in paths:
    path_obj = parse_path(p.attrib["d"])
    path_objs.append(path_obj)


html_str_1 = '''
<!DOCTYPE html>
<html>
<body>

<svg height="800" width="1000">
<g transform="scale(5)">
'''

html_path_str = ""

# for p in paths:
#     html_path_str += "<path d=\"" + p.attrib["d"] + "\" />\n"
        
# [cps for curve1, cps for curve2, ...]
paths_cps = []
# [length of curve1, length of curve2, ...]
paths_lengths = []
# [[curve1_id, curve2_id, ..], [curve1_id, curve2_id, ..], ...] 
shape_curve_ids = []

c_id = 0
for p_obj in path_objs:
    path_cps = []
    path_lens = []
    curve_ids = []
    for curve in p_obj:
        p_name = type(curve).__name__
        d_str = ""
        if p_name == "CubicBezier":
            cpts = np.zeros([4, 2])
            cpts[0][0] = curve.start.real
            cpts[0][1] = -curve.start.imag
            cpts[1][0] = curve.control1.real
            cpts[1][1] = -curve.control1.imag
            cpts[2][0] = curve.control2.real
            cpts[2][1] = -curve.control2.imag
            cpts[3][0] = curve.end.real
            cpts[3][1] = -curve.end.imag
            
            d_str = d_str + "M " + str(curve.start.real) + " " + str(curve.start.imag) + " "
            d_str = d_str + "C " + str(curve.control1.real) + " " + str(curve.control1.imag)
            d_str = d_str + ", " + str(curve.control2.real) + " " + str(curve.control2.imag)
            d_str = d_str + ", " + str(curve.end.real) + " " + str(curve.end.imag)
            
            path_cps.append(cpts)
            curve_ids.append(c_id)
        elif p_name == "Line" or p_name=="Close":
            if curve.length() == 0:
                continue
            cpts = np.zeros([2, 2])
            cpts[0][0] = curve.start.real
            cpts[0][1] = -curve.start.imag
            cpts[1][0] = curve.end.real
            cpts[1][1] = -curve.end.imag
        
            d_str = d_str + "M " + str(curve.start.real) + " " + str(curve.start.imag) + " "
            d_str = d_str + "L " + str(curve.end.real) + " " + str(curve.end.imag)
            
            path_cps.append(cpts)
            curve_ids.append(c_id)
        else:
            continue
            
        path_lens.append(curve.length())
        
        html_path_str += "<path id = \"" + str(c_id) + "\" d=\"" + d_str + "\" stroke=\"black\" fill=\"transparent\" />\n"
        c_id +=1
        
    shape_curve_ids.append(curve_ids)
    paths_cps.extend(path_cps)
    paths_lengths.extend(path_lens)
        
# label the paths
html_text_str = "<text font-size=\"4\">"

for i in range(len(paths_cps)):
    html_text_str = html_text_str + "<textPath href=\"#" + str(i) + "\" startOffset=\"40%\">" + str(i) + "</textPath>\n"
html_text_str += "</text>"

# end of html file
html_str_2 = '''
</g>
</svg>

</body>
</html>
'''
html_file = open("svg_display.html", "w")
html_file.write(html_str_1)
html_file.write(html_path_str)
html_file.write(html_text_str)
html_file.write(html_str_2)
html_file.close()
# print(shapes.attrib)
# path = parse_path('m 29.104167,146.1875 c 2.595859,1.42945 5.291666,1.32292 7.9375,0 l 2.645833,7.9375 c -3.96875,1.32292 -9.260417,1.32292 -13.229167,0 z')

# path
```

# specify pair of curves to match, if same orientation

```python
# path_pairs = [(0, 1, False), (3, 4, False), (5, 2, False), (6, 13, False), (7, 10, False), (8, 9, False), (11, 12, False)]
# path_pairs = [(0, 18, False), (1, 4, False), (2, 8, False), (3, 20, False), (5, 17, False), (6, 13, False), (7, 9, False), \
#     (10, 12, False), (11, 21, False), (14, 16, False), (15, 22, False), (19, 23, False)]
# path_pairs = [(0, 4, False)]
# path_pairs = [(0, 8, True), (1, 9, True), (4, 12, True), (5, 13, True), (14, 15, False), (10, 11, False), (2, 3, False), (6, 7, False)]
# path_pairs = [(6, 16, True), (7, 17, True), (1, 11, True), (2, 12, True), (10, 15, True), \
#               (14, 19, True), (13, 18, True), (5, 0, True), (9, 4, True), (8, 3, True)]
path_pairs = [(1, 13, True), (2, 14, True), (3, 15, True), (4, 16, True), (7, 19, True), (8, 20, True), (9, 21, True), (10, 22, True), \
                  (0, 11, False), (12, 23, False), (5, 6, False), (17, 18, False)]
```

```python
import math

path_lens = np.array(path_lens)
# at least have this many pts on one curve
min_curve_pts_num = 6
# desired segment length
seg_len = np.amin(path_lens) / min_curve_pts_num

# get number of vertices for each curve
curve_vert_nums = [0] * len(paths_cps)

for pair in path_pairs:
    # use the maximum curve length of the pair of curves
    c1_len = paths_lengths[pair[0]]
    c2_len = paths_lengths[pair[1]]
    max_len = max(c1_len, c2_len)
    # ensure that each segment is less than seg_len
    c_vert_num = math.ceil( max_len / seg_len)
    # set vertices num for both curves
    curve_vert_nums[pair[0]] = c_vert_num
    curve_vert_nums[pair[1]] = c_vert_num
    
for cid in range(len(paths_cps)):
    # skip if already set
    if curve_vert_nums[cid] == 0:
        c_len = paths_lengths[cid]
        curve_vert_nums[cid] = math.ceil( c_len / seg_len)
```

```python
bndry_verts = [None] * len(paths_cps)
err_thresh = 0.01
max_it = 20

def get_pair(c_id):
    global path_pairs
    for pair in path_pairs:
        if c_id == pair[0]:
            return pair[1]
        elif c_id == pair[1]:
            return pair[0]
    return -1

def getPtOnCurve(cps, t):
    return (1-t)**3 * cps[0] + 3*t*(1-t)**2 * cps[1] + 3*(1-t)*t**2 * cps[2] + t**3 * cps[3]

def getEquitDistantPtOnCurve(cps, pt_num, curve_len):
    N = 500
    prev_pt = cps[0]
    seg_lens = np.zeros(N)
    for i in range(1, N):
        t = i / N
        pt = getPtOnCurve(cps, t)
        seg_lens[i] = np.linalg.norm(pt - prev_pt)
        prev_pt = pt
    avg_seg_len = curve_len / pt_num
    bndry_pts = np.zeros([pt_num, 2])
    bndry_pts[0] = cps[0]
    
    last_seg = 0
    last_seg_part_len = 0
    ts = np.zeros(pt_num)
    
    for pt_id in range(1, pt_num):
        # find the segment up to which has avg_seg_len
        acc_len = last_seg_part_len
        for i in range(last_seg, N):
            if acc_len + seg_lens[i] < avg_seg_len:
                acc_len += seg_lens[i]
            else:
                # find the portion in segment
                last_seg_part_len = acc_len + seg_lens[i] - avg_seg_len
                last_seg = i + 1
                t = i / N - ( last_seg_part_len / seg_lens[i] ) / N
                ts[pt_id] = t
                bndry_pts[pt_id] = getPtOnCurve(cps, t)
#                 print("seg len = ", np.linalg.norm(bndry_pts[pt_id] - bndry_pts[pt_id-1]))
                break
    return bndry_pts
        
for curve_id in range(len(paths_cps)):
    curve_cps = paths_cps[curve_id]
    # number of verts on curve already calculated and stored in curve_vert_nums
    bndry_vert_num = curve_vert_nums[curve_id]
    # store curve vertices here
    curve_verts = np.zeros([bndry_vert_num, 2])
    
    # bezier curve
    if (len(curve_cps) == 4):
        curve_verts = getEquitDistantPtOnCurve(curve_cps, bndry_vert_num, paths_lengths[curve_id])
    # line
    elif (len(curve_cps) == 2):
        avg_seg_len = paths_lengths[curve_id] / bndry_vert_num    
        d_t = 1 / bndry_vert_num
        t = 0
        dir_vec = (curve_cps[1] - curve_cps[0])
        for v_id in range(bndry_vert_num):
            pt_loc = curve_cps[0] + t * dir_vec
            curve_verts[v_id] = pt_loc
            t += d_t
    bndry_verts[curve_id] = curve_verts
```

# displace matching boundaries

```python
# initially uv are copies of xy coords
bndry_vert_uv = np.array(bndry_verts_flat)

# displacement between paired boundaries
pair_bndry_dsplcmnt = []

# checkerboard transforms, (swap_xy, [x_coeff, y_coeff])
# inverse function are identical
trsfrm_funcs = [(False, np.array([1, 1])), (True, np.array([-1, 1])), [False, np.array([-1, -1])]]

def getNextCurveInShape(c_id):
    # find the shape that contains c_id
    global shape_curve_ids
    for shape_id in range(len(shape_curve_ids)):
        if c_id in shape_curve_ids[shape_id]:
            c_in_shape_idx = shape_curve_ids[shape_id].index(c_id)
            return shape_curve_ids[shape_id][(c_in_shape_idx + 1) % len(shape_curve_ids[shape_id])]
        
def getPrevCurveInShape(c_id):
    # find the shape that contains c_id
    global shape_curve_ids
    for shape_id in range(len(shape_curve_ids)):
        if c_id in shape_curve_ids[shape_id]:
            c_in_shape_idx = shape_curve_ids[shape_id].index(c_id)
            return shape_curve_ids[shape_id][(c_in_shape_idx - 1 + len(shape_curve_ids[shape_id])) % len(shape_curve_ids[shape_id])]

def applyTrnsfrm(coord, func):
    trnsfrmd_coord = np.zeros(2)
    if func[0]:
        trnsfrmd_coord[0] = coord[1]
        trnsfrmd_coord[1] = coord[0]
    else:
        trnsfrmd_coord[0] = coord[0]
        trnsfrmd_coord[1] = coord[1]
    trnsfrmd_coord = np.multiply(trnsfrmd_coord, func[1])
    return trnsfrmd_coord
    
def getDsplcmentArray(curve_vid_1, curve_vid_2, trnsfrm_func) :
    global bndry_vert_uv
    
    curve_verts_num = len(curve_vid_1)
    dsplcmentArray = np.zeros([curve_verts_num, 2])
    
    for i in range(curve_verts_num):
#         print ("before: ", bndry_vert_uv[curve_vid_1[i]])
#         print("after: ", applyTrnsfrm(bndry_vert_uv[curve_vid_1[i]], trnsfrm_func))
        dsplcmentArray[i] = applyTrnsfrm(bndry_vert_uv[curve_vid_1[i]], trnsfrm_func) - bndry_vert_uv[curve_vid_2[i]]

    return dsplcmentArray
    
    
def getPairCurveVids(pair_info):
    global bndry_c_verts_start, paths_lengths
    c_id1 = pair_info[0]
    c_id2 = pair_info[1]
    curve_vid_1 = np.arange(bndry_c_verts_start[c_id1], bndry_c_verts_start[c_id1] + len(bndry_verts[c_id1]))
    curve_vid_2 = np.arange(bndry_c_verts_start[c_id2], bndry_c_verts_start[c_id2] + len(bndry_verts[c_id2]))
    curve_vid_1 = np.append(curve_vid_1, bndry_c_verts_start[getNextCurveInShape(c_id1)]).astype(int)
    curve_vid_2 = np.append(curve_vid_2, bndry_c_verts_start[getNextCurveInShape(c_id2)]).astype(int)
    
    if not (pair_info[2]):
        curve_vid_2 = np.flip(curve_vid_2)
        
    return (curve_vid_1, curve_vid_2)

def getDsplcmntAdjSumSquare(pair_info, trnsfrm_func_id):
    global trsfrm_funcs
    trnsfrm_func = trsfrm_funcs[trnsfrm_func_id]
    curve_vid_1, curve_vid_2 = getPairCurveVids(pair_info)
#     print("pair_info: ", pair_info)
#     print("curve_vid_1: ", curve_vid_1)
#     print("curve_vid_2: ", curve_vid_2)
    curve_verts_num = len(curve_vid_1)
    
    dsplcmentArray = getDsplcmentArray(curve_vid_1, curve_vid_2, trnsfrm_func)
#     print("dsplcmentArray: ", dsplcmentArray)
    # calculate average displacement
    avg_displcmt = np.sum(dsplcmentArray, axis=0) / curve_verts_num
    # displacment has to be integers
    avg_displcmt = np.round(avg_displcmt)
    
    if trnsfrm_func_id == 1:
        avg_displcmt[0] += 0.5
        
    dsplcmntAdjArray = dsplcmentArray - avg_displcmt
    dsplcmntSum = np.sum(np.square(dsplcmntAdjArray))
    
    return (avg_displcmt, dsplcmntSum)
    
    
# find matching curve adjustments
for path_pair in path_pairs:
    avg_displcmt = np.zeros(2)

    avg_displcmt_min = np.zeros(2)
    dsplcmntSum_min = np.Inf
    trnsfrm_func_id = -1
    
    # try each transfer function
    for i in range(len(trsfrm_funcs)):
#     for i in [1]:
        avg_displcmt, dsplcmntSum = getDsplcmntAdjSumSquare(path_pair, i)
        print("func: " , i)
        print("avg_displcmt: ", avg_displcmt)
        print ("dsplcmntSum: ", dsplcmntSum)
        if dsplcmntSum < dsplcmntSum_min:
            dsplcmntSum_min = dsplcmntSum
            avg_displcmt_min = avg_displcmt
            trnsfrm_func_id = i

    pair_bndry_dsplcmnt.append([avg_displcmt_min, trnsfrm_func_id])

for i in range(len(pair_bndry_dsplcmnt)):    
    print(path_pairs[i], ": ", pair_bndry_dsplcmnt[i])
```

# group variables

```python
# endpoint v_id to cid map
explcit_ep_ids = {}

variable_ids = {}

def getCurveEnd(cid):
    global bndry_c_verts_start
    next_cid = getNextCurveInShape(cid)
    return bndry_c_verts_start[next_cid]

def getEPs(cid):
    global bndry_c_verts_start
    return [bndry_c_verts_start[cid], getCurveEnd(cid)]
      
def swap(arr):
    return [arr[1], arr[0]]

# x, y are separate vars
cnt = 0
for pair_id in range(len(path_pairs)):
    path_pair = path_pairs[pair_id]
    for i in range(2):
        eps = getEPs(path_pair[i])
        for j in range(2):
            if not(eps[j] in variable_ids):
                variable_ids[eps[j]] = cnt
                cnt += 2

def getVarIds(vids):
    global variable_ids
    var_ids = [0] * len(vids)
    for i in range(len(vids)):
        var_ids[i] = variable_ids[vids[i]]
    return var_ids

def getRelationFunc(ep, ep_pair, dsplcmnt_info):
    global bndry_verts_flat
    rel_funcs = np.zeros([2,2])
    trnsfrm_id = dsplcmnt_info[1]
    
    # same xy
    if trnsfrm_id == 0:
        trslate_funcs = np.zeros(2)
        for dim in range(2):
            trslate_funcs[dim] = - ( dsplcmnt_info[0][dim] - bndry_verts_flat[ep][dim] + bndry_verts_flat[ep_pair][dim] )
        # for x
        rel_funcs[0] = [1, trslate_funcs[0]]
        # for y
        rel_funcs[1] = [1, trslate_funcs[1]]

    # eps flip xy
    elif trnsfrm_id == 1:
        trslate_funcs = np.zeros(2)
        trslate_funcs[0] = - ( dsplcmnt_info[0][0] - bndry_verts_flat[ep][1] + bndry_verts_flat[ep_pair][0])
        trslate_funcs[1] = - ( dsplcmnt_info[0][1] - bndry_verts_flat[ep][0] + bndry_verts_flat[ep_pair][1])
        # for x
        rel_funcs[0] = [1, trslate_funcs[0]]
        # for y
        rel_funcs[1] = [1, trslate_funcs[1]]

    # eps flip xy, *-1
    elif trnsfrm_id == 2:
        trslate_funcs = np.zeros(2)
        trslate_funcs[0] = - ( dsplcmnt_info[0][0] + bndry_verts_flat[ep][1] + bndry_verts_flat[ep_pair][0])
        trslate_funcs[1] = - ( dsplcmnt_info[0][1] + bndry_verts_flat[ep][0] + bndry_verts_flat[ep_pair][1])
        # for x
        rel_funcs[0] = [-1, trslate_funcs[0]]
        # for y
        rel_funcs[1] = [-1, trslate_funcs[1]]

    # same xy, *-1
    elif trnsfrm_id == 3:
        trslate_funcs = np.zeros(2)
        for dim in range(2):
            trslate_funcs[dim] = - ( dsplcmnt_info[0][dim] + bndry_verts_flat[ep][dim] + bndry_verts_flat[ep_pair][dim] )
        # for x
        rel_funcs[0] = [-1, trslate_funcs[0]]
        # for y
        rel_funcs[1] = [-1, trslate_funcs[1]]
    return rel_funcs
    
# func1 to be put into func2
def compose_rel_func(func1, func2):
    return [func1[0] * func2[0], func2[0] * func1[1] + func2[1]]

def other_dim_id(var_id):
    if var_id % 2 == 1:
        return var_id - 1
    return var_id + 1

def add_rel_vars(ep, ep_pair, rel_funcs, trnsfrm_id):
    global explcit_ep_ids
    # no flip
    if trnsfrm_id == 0 or trnsfrm_id == 3:
        # for x
        explcit_ep_ids[ep].append((ep_pair, rel_funcs[0]))
        # for y
        explcit_ep_ids[ep+1].append((ep_pair+1, rel_funcs[1]))
    # flip xy
    else:
        # for x
        explcit_ep_ids[ep+1].append((ep_pair, rel_funcs[0]))
        # for y
        explcit_ep_ids[ep].append((ep_pair+1, rel_funcs[1]))

# group related vars
var_groups = []

def get_group_id(var_id):
    global var_groups
    for i in range(len(var_groups)):
        if var_id in var_groups[i]:
            return i
    return -1

def get_trnsfrm_id(pair_id):
    global pair_bndry_dsplcmnt
    dsplcmnt_info = pair_bndry_dsplcmnt[pair_id]
    trnsfrm_id = dsplcmnt_info[1]
    return trnsfrm_id

def get_ep_pair(pair_id):
    global path_pairs
    
    path_pair = path_pairs[pair_id]
    
    eps = getEPs(path_pair[0])    
    ep_pairs = getEPs(path_pair[1])
    if not (path_pair[2]):
        ep_pairs = swap(ep_pairs)
    
    return (eps, ep_pairs)

def get_var_pair(pair_id):

    eps, ep_pairs = get_ep_pair(pair_id)
    ep_var_ids = getVarIds(eps)
    eppair_var_ids = getVarIds(ep_pairs)
    trnsfrm_id = get_trnsfrm_id(pair_id)
    
    var_pairs = [None] * 2
    
    for i in range(2):
        x1_id = ep_var_ids[i]
        y1_id = x1_id + 1
        x2_id = eppair_var_ids[i]
        y2_id = x2_id + 1
        
        if trnsfrm_id == 0 or trnsfrm_id == 3:
            var_pairs[i] = [[x1_id, x2_id], [y1_id, y2_id]]
        else:
            var_pairs[i] = [[x1_id, y1_id], [x2_id, y2_id]]
    return var_pairs
       
pair_func_to_be_fixed = set()
var_pair_no_existent = []
var_pair_to_be_fixed = []

# group related vars         
for pair_id in range(len(path_pairs)):
    var_pairs = get_var_pair(pair_id)

    # start and end
    for i in range(2):
        pairs = var_pairs[i]
        
        # two dim
        for j in range(2):
            g1_id = get_group_id(pairs[j][0])
            g2_id = get_group_id(pairs[j][1])
            
            # merge var_pair_id to g1
            if g1_id >=0 and g2_id < 0:
                var_groups[g1_id].append(pairs[j][1])
            # merge var_id to g2
            elif g2_id >=0 and g1_id < 0:
                var_groups[g2_id].append(pairs[j][0])
            # both var are in groups
            elif g1_id >=0 and g2_id >= 0:
                # already in same group, a circle
                if g1_id == g2_id:
                    # remove circle, change boundry relation later
                    pair_func_to_be_fixed.add(pair_id)
                    # do not find relationship between the two vars
                    var_pair_no_existent.append(pairs[j])
                    # get other endpoint
                    if i == 0:
                        other_pairs = var_pairs[1]
                    else:
                        other_pairs = var_pairs[0]
                    var_pair_to_be_fixed.append(other_pairs[j])
                    continue
                # in different group, merge them
                else:
                    var_groups[g1_id] = var_groups[g1_id] + var_groups[g2_id]
                    var_groups.pop(g2_id) 
            # no relevant group exist
            else:
                var_groups.append(pairs[j])
    
# var_pair_no_existent
# var_pair_to_be_fixed
for var_pair in var_pair_to_be_fixed:
    for i in range(len(var_groups)):
        var_group = var_groups[i]
        if var_pair[0] in var_group :
                var_groups.pop(i)
                break

variable_ids
# var_groups
# var_pair_to_be_fixed
```

# find pairwise relationships

```python
# find pairwise relationships
pair_rel_funcs = {}

for pair_id in range(len(path_pairs)):
    eps, ep_pairs = get_ep_pair(pair_id)
    var_pairs = get_var_pair(pair_id)
    dsplcmnt_info = pair_bndry_dsplcmnt[pair_id]
    
    # start and end v
    for i in range(2):
        pairs = var_pairs[i]
        
        # if the pair creates a circle, skip
        if pairs[0] in var_pair_no_existent:
            continue
            
        if pairs[0] in var_pair_to_be_fixed:
            continue
        rel_funcs = getRelationFunc(eps[i], ep_pairs[i], dsplcmnt_info)
        
        # x and y
        for j in range(2):
            pair_rel_funcs[(pairs[j][0], pairs[j][1])] = rel_funcs[j]
            
pair_rel_funcs
```

# find relationship between the first of the group to the rest of a group

```python
def inv_rel_func(func):
    return [1 / func[0], -func[1] / func[0]]

# find relationship between the first of the group to the rest of a group, do search
for group in var_groups:
    # first var id in group
    cur_var = group[0]
    # create an entry in explcit_ep_ids
    explcit_ep_ids[cur_var] = {}
    
    next_vars = []
    for i in range(1, len(group)):
        rel_var = group[i]
        # go through all pairs
        for var_pair in pair_rel_funcs:
            # directly related
            if cur_var in var_pair and rel_var in var_pair:
                # add to next vars to process
                next_vars.append(rel_var)
                # no need to invert rel_func
                if var_pair.index(rel_var) == 1:
                    explcit_ep_ids[group[0]][rel_var] = pair_rel_funcs[var_pair]
                else:
                    explcit_ep_ids[group[0]][rel_var] = inv_rel_func(pair_rel_funcs[var_pair])
                    
    done_vars = [cur_var]
    
    while len(next_vars) > 0:
        cur_var = next_vars.pop(0)
        # go through all pairs
        for var_pair in pair_rel_funcs:
            if cur_var in var_pair:
                if var_pair.index(cur_var) == 0:
                    rel_var = var_pair[1]
                else:
                    rel_var = var_pair[0]
                    
                # already processed this pair
                if rel_var in done_vars:
                    continue
                else:
                    # process rel_var later
                    next_vars.append(rel_var)
                    if var_pair.index(cur_var) == 0:
                        rel_func = pair_rel_funcs[var_pair]
                    else:
                        rel_func = inv_rel_func(pair_rel_funcs[var_pair])
                    
                        
                    # compose function
                    cmpsd_func = compose_rel_func(explcit_ep_ids[group[0]][cur_var], rel_func)
                    explcit_ep_ids[group[0]][rel_var] = cmpsd_func
        # done with cur_var
        done_vars.append(cur_var)
        
explcit_ep_ids
```

# find relationship between circle variable pairs

```python
rel_func_var_pair_no_existent = {}

for var_pair in var_pair_no_existent :
    # get the group
    for key_var in explcit_ep_ids:
        # if one of the pair var is key
        if var_pair[0] == key_var:
            rel_func = explcit_ep_ids[key_var][var_pair[1]]
            rel_func_var_pair_no_existent[tuple(var_pair)] = rel_func
        elif var_pair[1] == key_var:
            rel_func = inv_rel_func(explcit_ep_ids[key_var][var_pair[0]])
            rel_func_var_pair_no_existent[tuple(var_pair)] = rel_func
        elif var_pair[0] in explcit_ep_ids[key_var]:
            func1 = explcit_ep_ids[key_var][var_pair[0]]
            func2 = explcit_ep_ids[key_var][var_pair[1]]
            inv_func1 = inv_rel_func(func1)
            rel_func = compose_rel_func(inv_func1, func2)
        
            rel_func_var_pair_no_existent[tuple(var_pair)] = rel_func
            
rel_func_var_pair_no_existent            
# var_pair_no_existent
```

# calculate values

```python
var_values = {}
for v_id in variable_ids:
    var_values[variable_ids[v_id]] = 0
    var_values[variable_ids[v_id] + 1] = 0

def cal_vars(explcit_ep_ids):
    global var_values
    
    # calculate the explicit vars
    for key in explcit_ep_ids:
        coeff_1 = len(explcit_ep_ids[key]) + 1
        coeff_2 = 0
        for rel_var in explcit_ep_ids[key]:
            coeff_2 += explcit_ep_ids[key][rel_var][1] * explcit_ep_ids[key][rel_var][0]
        var_value = - coeff_2 / coeff_1
        var_values[key] = var_value

    # calculate the implicit vars
    for key in explcit_ep_ids:
        for rel_var in explcit_ep_ids[key]:
            var_value = var_values[key] *  explcit_ep_ids[key][rel_var][0] + explcit_ep_ids[key][rel_var][1]
            var_values[rel_var] = var_value

cal_vars(explcit_ep_ids)            
var_values
```

# find relationship between to-be-defined variable pairs

```python
def getVid(var_id):
    if var_id % 2 == 1:
        var_id -= 1
    global variable_ids
    for key in variable_ids:
        if variable_ids[key] == var_id:
            return key

def getVarVidValue(var_id):
    global bndry_verts_flat
    vid = getVid(var_id)
    if var_id % 2 == 1:
        return bndry_verts_flat[vid][1]
    return bndry_verts_flat[vid][0]
    
def get2BfixedPairId(var_pair):
    global pair_func_to_be_fixed
    for pair_id in pair_func_to_be_fixed:
        var_pairs = get_var_pair(pair_id)
        for i in range(2):
            for j in range(2):
                if var_pair[0] in var_pairs[i][j] and var_pair[1] in var_pairs[i][j]:
                    return pair_id

    
snd_explcit_ep_ids = {}

# for pair_id in pair_func_to_be_fixed:
#     var_pairs = get_var_pair(pair_id)
#     for i in range(2):
#         for dim in range(2):
#             var_pair = tuple(var_pairs[i][dim])
            
#             # recalculate bndry transform function
#             if var_pair in rel_func_var_pair_no_existent:
#                 rel_func = rel_func_var_pair_no_existent[var_pair]
#                 loc_val_1 = getVarVidValue(var_pair[0])
#                 loc_val_2 = getVarVidValue(var_pair[1])
#                 if rel_func[0] == 1:
#                     translate = - rel_func[1] + loc_val_1 - loc_val_2
#                     pair_bndry_dsplcmnt[pair_id][1] = 1
#                 elif rel_func[0] == -1:
#                     translate = - rel_func[1] - loc_val_1 - loc_val_2
#                     pair_bndry_dsplcmnt[pair_id][1] = 3
# #                 print("translate: ", translate)
#                 if var_pair[0] % 2 == 0:
#                     pair_bndry_dsplcmnt[pair_id][0][0] = translate
#                 else:
#                     pair_bndry_dsplcmnt[pair_id][0][1] = translate

for pair_id in pair_func_to_be_fixed:
    var_pairs = get_var_pair(pair_id)
    pair_info = path_pairs[pair_id]
    
    for i in range(2):      
        # recalculate bndry transform function
        if tuple(var_pairs[i][0]) in rel_func_var_pair_no_existent:
            var_pair = var_pairs[i]
            loc_val = np.zeros([2,2])
    
            for se in range(2):
                for dim in range(2):
                    loc_val[se][dim] = getVarVidValue(var_pair[dim][se])
                    loc_val[se][dim] += var_values[var_pair[dim][se]]

            translate1 = loc_val[0] - loc_val[1]
            translate2 = -loc_val[0] - loc_val[1]
            print("translate1: ", translate1)
            print("translate2: ", translate2)
            curve_vid_1, curve_vid_2 = getPairCurveVids(pair_info)

            # try 1st and 4th transform function
            dsplcmentArray_1 = getDsplcmentArray(curve_vid_1, curve_vid_2, trsfrm_funcs[0])[1:]
            dsplcmntAdjArray_1 = dsplcmentArray_1 - translate1
            dsplcmntSum_1 = np.sum(np.square(dsplcmntAdjArray_1))
            
            dsplcmentArray_2 = getDsplcmentArray(curve_vid_1, curve_vid_2, trsfrm_funcs[3])[1:]
            dsplcmntAdjArray_2 = dsplcmentArray_2 - translate2
            dsplcmntSum_2 = np.sum(np.square(dsplcmntAdjArray_2))
            
            if dsplcmntSum_1 < dsplcmntSum_2:
                pair_bndry_dsplcmnt[pair_id] = [translate1, 0]
            else:
                pair_bndry_dsplcmnt[pair_id] = [translate2, 3]

            
for var_pair in var_pair_to_be_fixed:
    vid_1 = getVid(var_pair[0])
    vid_2 = getVid(var_pair[1])
    pair_id = get2BfixedPairId(var_pair)
#     print("var_pair: ", var_pair)
#     print("pair_id: ", pair_id)
#     print("pair_bndry_dsplcmnt[pair_id]: ", pair_bndry_dsplcmnt[pair_id])
    rel_funcs = getRelationFunc(vid_1, vid_2, pair_bndry_dsplcmnt[pair_id])
    # x
    if var_pair[0] % 2 == 0:
        snd_explcit_ep_ids[var_pair[0]] = {var_pair[1]: rel_funcs[0]}
    else:
        snd_explcit_ep_ids[var_pair[0]] = {var_pair[1]: rel_funcs[1]}
        
# snd_explcit_ep_ids

cal_vars(snd_explcit_ep_ids)
# var_values
```

```python
bndry_vert_uv = np.copy(bndry_verts_flat)

# move the endpoints first
for vid in variable_ids:
    bndry_vert_uv[vid][0] = bndry_verts_flat[vid][0] + var_values[variable_ids[vid]]
    bndry_vert_uv[vid][1] = bndry_verts_flat[vid][1] + var_values[variable_ids[vid]+1]

def intrp_curve(cid):
    global bndry_vert_uv, bndry_verts_flat
    start_vid, end_vid = getEPs(cid)
        
    end_dsplcmnt = bndry_vert_uv[end_vid] - bndry_verts_flat[end_vid]
    start_dsplcmnt = bndry_vert_uv[start_vid] - bndry_verts_flat[start_vid]
    curve_v_num = len(bndry_verts[cid])
    
    for i in range(1, curve_v_num):
        v_dsplcmnt = ( i / curve_v_num ) * end_dsplcmnt + ( 1 - i / curve_v_num ) * start_dsplcmnt
#         print("added: ", v_dsplcmnt)
        bndry_vert_uv[start_vid + i] = bndry_verts_flat[start_vid + i] + v_dsplcmnt
        
        
# move the curve pairs
for pair_id in range(len(path_pairs)):
#     eps, ep_pairs = get_ep_pair(pair_id)
    eps = getEPs(path_pairs[pair_id][0])
    curve_vid_1, curve_vid_2 = getPairCurveVids(path_pairs[pair_id])
    c1_len = len(curve_vid_1)
    
    var_pairs = get_var_pair(pair_id)
    dsplcmnt_info = pair_bndry_dsplcmnt[pair_id]
    
    intrp_curve(path_pairs[pair_id][0])
        
    # use rel transform + translate to get second curve
    for t in range(1, c1_len-1):
        c1_vid = curve_vid_1[t]
        c2_vid = curve_vid_2[t]
        bndry_vert_uv[c2_vid] = applyTrnsfrm(bndry_vert_uv[c1_vid], trsfrm_funcs[dsplcmnt_info[1]]) - dsplcmnt_info[0]

# pair_func_to_be_fixed
for pair_id in pair_func_to_be_fixed:
    eps = getEPs(path_pairs[pair_id][0])
    curve_vid_1, curve_vid_2 = getPairCurveVids(path_pairs[pair_id])
    c1_len = len(curve_vid_1)
    
    var_pairs = get_var_pair(pair_id)
    dsplcmnt_info = pair_bndry_dsplcmnt[pair_id]
    
    intrp_curve(path_pairs[pair_id][0])
        
    # use rel transform + translate to get second curve
    for t in range(1, c1_len-1):
        c1_vid = curve_vid_1[t]
        c2_vid = curve_vid_2[t]
        bndry_vert_uv[c2_vid] = applyTrnsfrm(bndry_vert_uv[c1_vid], trsfrm_funcs[dsplcmnt_info[1]]) - dsplcmnt_info[0]
    
    
# gather indices of curves needed to be interpolated

cid_intrp = set()
cid_pair = set()

for path_pair in path_pairs:
    cid_pair.add(path_pair[0])
    cid_pair.add(path_pair[1])
    
for path_pair in path_pairs:
    for pid in range(2):
        next_cid = getNextCurveInShape(path_pair[pid])
        prev_cid = getPrevCurveInShape(path_pair[pid])
        if not (next_cid in cid_pair):
            cid_intrp.add(next_cid)
        if not (prev_cid in cid_pair):
            cid_intrp.add(prev_cid)
        
# do the interpolation
for cid in cid_intrp:
    intrp_curve(cid)
# bndry_vert_uv
```
